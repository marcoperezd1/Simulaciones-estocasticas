```python
import numpy as np
import matplotlib.pyplot as plt
```

Los incrementos brownianos están dados de la forma:

$$
\Delta W_i = W_{t_{i+1}}-  W_{t_{i}}, \qquad \Delta W_i = \sqrt{\Delta t}\ Z_i,
$$
donde estamos asumiento que cada incremento tiene el mismo largo.

```python
# Generamos una trayectoria browniana
T = 1.0
n = 1000
dt = T/n #partición x-espaciada

t = np.linspace(0, T, n + 1)

dW = np.sqrt(dt) * np.random.randn(n)

W = np.zeros(n + 1)
W[1:] = np.cumsum(dW)
#despues del 0, pq parte en 0
plt.plot(t, W, c = 'blue')
plt.title("Movimiento Browniano")
plt.xlabel("t")
plt.ylabel("W_t")
plt.show()
```

Ahora, calcularemos:
$$
\int_0^T W_t dW_t
$$

```python
integral_ito = np.sum(W[:-1]* dW)
print(f"\n La aproximación de la integral de Ito es: {integral_ito}")
```

Integremos un proceso simple adaptado, es decir:
$$
X_t = \sum_{i=0}^{n-1} \xi_i 1_{(t_{i}, t_{i+1}]}(t)$$
para tiempos
$$
0=t_0 < t_1 < \ldots < t_n = T
$$

Integraremos en este caso:

$$
X = 1\cdot 1_{(0,1]}(t) + 2 \cdot 1_{(1,2]}(t)+ 3 \cdot 1_{(2,3]}(t)
$$

```python
T = 3
n = 3
dt = 1

dW = np.sqrt(dt) * np.random.randn(n)

W = np.zeros(n + 1) #definimos la trayectoria brownianas
W[1:] = np.cumsum(dW)

X = np.array([1,2,3])

integral_ito = np.sum(X * dW)


print("dW:", dW)
print("W:", W)
print("Integral de Ito:", integral_ito)
```
